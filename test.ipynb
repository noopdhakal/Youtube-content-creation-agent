```python
import os
from dotenv import load_dotenv
import google.generativeai as genai
from tavily import TavilyClient

load_dotenv()

genai.configure(api_key=os.getenv("GEMINI_API_KEY"))
tavily_client = TavilyClient(api_key=os.getenv("TAVILY_API_KEY"))

def get_realtime_info(query):
    resp = tavily_client.search(query=query, max_results=3)

    summaries = []
    for r in resp["results"]:
        title = r.get("title", "")
        snippet = r.get("snippet", "")
        url = r.get("url", "")
        summaries.append(f"{title}\n{snippet}\n{url}")

    source_info = "\n\n---\n\n".join(summaries)

    model = genai.GenerativeModel("gemini-2.5-flash")

    prompt = f"Summarize this:\n{source_info}"

    response = model.generate_content(prompt)

    return response.text


print(get_realtime_info("AI in healthcare"))
```
